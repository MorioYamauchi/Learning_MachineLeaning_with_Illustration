# 機械学習とは

## 1.1 学習の種類
---
 - 教師付き学習
     - コンピュータに問題と答えの対を与えることにより，未知の問題にも正しく回答できる**汎化能力**をコンピュータに獲得させることが目標．
     - 実数値の答えを予測する回帰，カテゴリ値の答えを予測する分類，順序を予測するランキングなどが典型的なタスクである．
     - 応用例としては，手書き文字認識，音声認識，画像認識，迷惑メールの自動分類，ウェブページの検索，株価予測などがある．
 - 教師なし学習
     - コンピュータは教師信号なしで，自分一人でデータから情報を集め学習する．
     - データのクラスタリングや異常検出などが典型的なタスクとして挙げられる．
     - SNSの解析やデータの可視化，教師付き学習のための前処理として用いられることがある．
 - 強化学習
     - 教師付き学習と同じくコンピュータに**汎化能力**を獲得させることが目的だが，教師信号がないという点で異なる．強化学習では，教師信号の代わりとして，出力結果に対する評価が行われる．コンピュータは評価結果をもとに，より評価が高まるように学習を進める．
     - ロボットの自律制御やコンピュータゲームの人工知能，マーケティング戦略の最適化などに応用されている．

## 1.2 機械学習のタスクの例
---
 - 回帰
     - 実数値関数を標本点から近似する教師付きの関数近似問題．
 - 分類
     - 与えられたパターンを識別する教師付きのパターン認識問題．
     - 出力は実数値ではなくクラスラベルをとる．
     - クラスラベルを出力する関数を学習する問題と等価であるため，分類も回帰と同様に関数近似問題として捉えることができる．
 - 異常検出     
     - 与えられた入力標本に含まれる異常値を見つける問題．
     - 異常値の特徴が明示されている場合は教師付きの分類問題と等価である．明示されていない場合は，密集したデータを正常とみなし，そこから外れた場所に位置するデータを異常値とみなす．
 - クラスタリング
     - 分類と同じくパターン認識問題であるが，教師なしという点で異なる．
     - 類似するデータを同一のクラスに属するように学習を進めるため，標本間の類似度を適切に決定することが重要．
 - 次元削減
     - 高次元データから本質的な情報を抽出し，低次元の表現へと変換する問題．
     - 入出力が対になった標本が与えられた場合は教師付き次元削減となり，汎化能力の高い低次元表現を求める．
     - 入力標本だけが与えられるような教師なし次元削減では，データ分布の大域的な形状やデータ間の局所的な隣接関係など，入力標本の何らかの特性を保存する低次元表現を求める．
